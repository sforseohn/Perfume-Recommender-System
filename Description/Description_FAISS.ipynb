```python
import pandas as pd
import os

# ÎîîÎ†âÌÜ†Î¶¨ Í≤ΩÎ°ú
chypre_path = "/root/.cache/kagglehub/datasets/joehusseinmama/fragrantica-data/versions/1/chypre"

# Î™®Îì† ÌååÏùº Í∞ÄÏ†∏Ïò§Í∏∞
all_entries = os.listdir(chypre_path)

# Ïã§Ï†ú CSV ÌååÏùºÏù∏ Í≤ΩÏö∞Îßå ÌïÑÌÑ∞ÎßÅ (ÎîîÎ†âÌÜ†Î¶¨Îäî Ï†úÏô∏)
csv_files = [f for f in all_entries if f.endswith('.csv') and os.path.isfile(os.path.join(chypre_path, f))]

# Í∞Å ÌååÏùºÏùÑ pandasÎ°ú Î∂àÎü¨ÏôÄ ÏÉÅÏúÑ 10Í∞ú Ï∂úÎ†•
for file in csv_files:
    file_path = os.path.join(chypre_path, file)
    print(f"\nüìÑ {file} - ÏÉÅÏúÑ 10Í∞ú Ìñâ:")
    df = pd.read_csv(file_path)
    display(df.head(10))
```

## 1. description Îç∞Ïù¥ÌÑ∞ ÌôïÏù∏

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import faiss
import numpy as np

# 1. description ÌÖçÏä§Ìä∏ Ï∂îÏ∂ú
descriptions = df["description"].fillna("").tolist()
display(df[["title", "description"]].head(10))
```

## 2. Description TF-IDF Î≤°ÌÑ∞Î°ú Î≥ÄÌôò

```python
# TF-IDF Î≤°ÌÑ∞Ìôî
vectorizer = TfidfVectorizer(max_features=10000)
tfidf_matrix = vectorizer.fit_transform(descriptions)
tfidf_matrix
# (Ï†ÑÏ≤¥ Ìñ•Ïàò Í∞úÏàò, description Î≤°ÌÑ∞Î°ú Î∞îÍæº Ï∞®Ïõê Ïàò -> TF-IDFÍ∞Ä Î≤°ÌÑ∞ÌôîÌïú Îã®Ïñ¥ Ïàò)
```

## 3. FAISS Ïù∏Îç±Ïä§ Íµ¨Ï∂ï


*   TF-IDF Î≤°ÌÑ∞Î•º float32Î°ú Î≥ÄÌôò ÌõÑ
*   FAISSÏùò L2 Ïù∏Îç±Ïä§ (IndexFlatL2)Ïóê Ï†ÄÏû•
*   Í±∞Î¶¨(L2)Î•º Í∏∞Ï§ÄÏúºÎ°ú Ïú†ÏÇ¨Ìïú Ìñ•Ïàò Ï∞æÏùå





```python
# FAISSÎäî float32 ÌÉÄÏûÖÎßå ÏßÄÏõê
tfidf_vectors = tfidf_matrix.toarray().astype("float32")

# FAISS Ïù∏Îç±Ïä§ ÏÉùÏÑ± (L2 Í±∞Î¶¨)
index = faiss.IndexFlatL2(tfidf_vectors.shape[1])
index.add(tfidf_vectors)  # Ï†ÑÏ≤¥ Ìñ•Ïàò ÏÑ§Î™Ö Î≤°ÌÑ∞ Ï†ÄÏû•
```

## 4. ÏÇ¨Ïö©Ïûê ÏûÖÎ†•ÏùÑ Î∞îÌÉïÏúºÎ°ú Ïú†ÏÇ¨ Ìñ•Ïàò 5Í∞ú

```python
# ÏÇ¨Ïö©ÏûêÍ∞Ä Í≥†Î•∏ Ìñ•Ïàò Ïù¥Î¶Ñ
selected_name = "Good Girl"

# Ìï¥Îãπ Ìñ•ÏàòÏùò index Ï∞æÍ∏∞
selected_idx = df[df["title"].str.contains(selected_name, case=False, na=False)].index[0]

# Ìï¥Îãπ Ìñ•ÏàòÏùò TF-IDF Î≤°ÌÑ∞ Ï∂îÏ∂ú
query_vector = tfidf_vectors[selected_idx:selected_idx+1] # ÏûÖÎ†•Ìïú Ìñ•ÏàòÏùò description tf-idf Î≤°ÌÑ∞Ìôî

# FAISSÎ°ú Ïú†ÏÇ¨ Ìñ•Ïàò 5Í∞ú Í≤ÄÏÉâ
D, I = index.search(query_vector, k=5)

# Í≤∞Í≥º Ï∂úÎ†•
print("ÏÑ†ÌÉùÌïú Ìñ•Ïàò:", df.loc[selected_idx, "title"])
print("Ï∂îÏ≤ú Ìñ•Ïàò:")
display(df.iloc[I[0]][["title", "designer", "description"]])
```
